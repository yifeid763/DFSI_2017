<h2><font size = "4">Analysis 1</font></h2>
(TAB INTERACTION TEMPLATE)

```python
import pandas
from bokeh.models import *
from bokeh.plotting import *
from bokeh.io import show


f_agg = f.groupby('Year').mean().reset_index()

year = list()
for i in range(len(f_agg['Year'])):
    year.append(f_agg['Year'][i])

waterloss = list()
for i in range(len(f_agg['Actual waterloss in period (i.e actual evapotranspiration with no irrigation) mm/ha/month'])):
    waterloss.append(f_agg['Actual waterloss in period (i.e actual evapotranspiration with no irrigation) mm/ha/month'][i])
p1 = figure(plot_width = 600, plot_height = 400
           ,x_axis_label = 'Year'
           ,y_axis_label = 'Average Monthly Waterloss (mm/Ha)')
p1.vbar(x = year, top = waterloss, bottom = 0, width = 0.5, color = "#3f9eff")


deficit = list()
for i in range(len(f_agg['Relative Evapotranspiration Deficit (no irrigation) '])):
    deficit.append(f_agg['Relative Evapotranspiration Deficit (no irrigation) '][i])
p2 = figure(plot_width = 600, plot_height = 400
           ,x_axis_label = 'Year'
           ,y_axis_label = 'Relative ET Deficit')
p2.vbar(x = year, top = deficit, bottom = 0, width = 0.5, color = "#2E94AB")


et = list()
for i in range(len(f_agg['Adjusted actual evapotranspiration (with irrigation) mm/month'])):
    et.append(f_agg['Adjusted actual evapotranspiration (with irrigation) mm/month'][i])
p3 = figure(plot_width = 600, plot_height = 400
           ,x_axis_label = 'Year'
           ,y_axis_label = 'Actual ET (mm/month)')
p3.line(x = year, y = et, color = "#9d1db2", line_width = 1.8)



tab1 = Panel(child = p1, title = "Waterloss")
tab2 = Panel(child = p2, title = "ET Deficit")
tab3 = Panel(child = p3, title = "Actual ET")

tabs = Tabs(tabs = [tab1,tab2,tab3])

show(tabs)
```

These charts show the anuual waterloss, evapotranspiration(ET) deficit and actual ET:
1. Waterloss
   - Discussion 1
   - Discussion 2
2. ET Deficit
3. Actual ET



<h2><font size = "4">Analysis 2</font></h2>
(SCATTER PLOT INTERACTION TEMPLATE)

```python
from bokeh.palettes import Spectral

x1 = list()
y1 = list()
x2 = list()
y2 = list()
x3 = list()
y3 = list()
x4 = list()
y4 = list()
x5 = list()
y5 = list()

for i in range(len(f['Year'])):
    if f['Growth stage'][i] == '1. Emergence':
        x1.append(f['Actual waterloss in period (i.e actual evapotranspiration with no irrigation) mm/ha/month'][i])
        y1.append(f['Relative Evapotranspiration Deficit (no irrigation) '][i])
    elif f['Growth stage'][i] == '2. First Square':
        x2.append(f['Actual waterloss in period (i.e actual evapotranspiration with no irrigation) mm/ha/month'][i])
        y2.append(f['Relative Evapotranspiration Deficit (no irrigation) '][i])
    elif f['Growth stage'][i] == '3. First flower':
        x3.append(f['Actual waterloss in period (i.e actual evapotranspiration with no irrigation) mm/ha/month'][i])
        y3.append(f['Relative Evapotranspiration Deficit (no irrigation) '][i])
    elif f['Growth stage'][i] == '4. Flowering':
        x4.append(f['Actual waterloss in period (i.e actual evapotranspiration with no irrigation) mm/ha/month'][i])
        y4.append(f['Relative Evapotranspiration Deficit (no irrigation) '][i])
    else:
        x5.append(f['Actual waterloss in period (i.e actual evapotranspiration with no irrigation) mm/ha/month'][i])
        y5.append(f['Relative Evapotranspiration Deficit (no irrigation) '][i])

         
hover = HoverTool(
        tooltips=[
            ("Waterloss", "$x"),
            ("ET Deficit", "$y")
        ]
    )        

TOOLS="pan,wheel_zoom,reset,hover,poly_select,box_select"   

ps = figure(plot_width = 500, plot_height = 500
           ,x_axis_label = 'Waterloss'
           ,y_axis_label = 'ET Deficit'
           ,tools = TOOLS)

ps.circle(x1, y1, color=Spectral[6][1],line_width=5,legend = '1. Emergence')
ps.circle(x2, y2, color=Spectral[6][2],line_width=5,legend = '2. First Square')
ps.circle(x3, y3, color=Spectral[6][3],line_width=5,legend = '3. First flower')
ps.circle(x4, y4, color=Spectral[6][4],line_width=5,legend = '4. Flowering')
ps.circle(x5, y5, color=Spectral[6][5],line_width=5,legend = '5. Boll Development')

ps.legend.location = "bottom_left"

show(ps)
```

This scatter plot illustrates the relationship between ET Deficit and Waterloss
